# Phase 3 — Data Cleaning

This notebook merges `Fake.csv` and `True.csv` under `data/raw/` into a single dataset, shuffles it, combines `title` and `text`, removes duplicates, and saves it to `data/processed/cleaned.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load raw datasets
fake_df = pd.read_csv("../data/raw/Fake.csv")
true_df = pd.read_csv("../data/raw/True.csv")

fake_df["label"] = 0
true_df["label"] = 1

df = pd.concat([fake_df, true_df], ignore_index=True)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

print(f"Total articles loaded: {len(df):,}")
print(df["label"].value_counts())
```

```python
# Combine title and text
df["content"] = df["title"].fillna("") + " " + df["text"].fillna("")

# Drop duplicates
before = len(df)
df.drop_duplicates(subset=["content"], inplace=True)
print(f"Dropped {before - len(df):,} duplicate articles")
```
